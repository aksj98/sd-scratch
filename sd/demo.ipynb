```python
import model_loader
import pipeline
from PIL import Image
from transformers import CLIPTokenizer
import torch
```

```python
DEVICE="cpu"

ALLOW_CUDA=False
ALLOW_MPS=False

if ALLOW_CUDA and torch.cuda.is_available():
    DEVICE="cuda"
elif ALLOW_MPS and torch.backends.mps.is_available():
    DEVICE="mps"

print(f"Using device: {DEVICE}")

tokenizer=CLIPTokenizer("../data/vocab.json",merges_file="../data/merges.txt")
model_file="../data/v1-5-pruned-emaonly.ckpt"

models=model_loader.preload_models_from_standard_weights(model_file,DEVICE)

## Text to image 
prompt="A gorilla in the jungle"
uncond_prompt="beaches"
do_cfg=True
cfg_scale=7.5
```

```python
input_image=None
# input_image=Image.open("../images/dog.jpg")
strength=0.9

sampler="ddpm"
num_inference_steps=50
seed=42

image=pipeline.generate(
    prompt,uncond_prompt,input_image=input_image,
    n_inference_steps=num_inference_steps,models=models,seed=seed,device=DEVICE,idle_device=DEVICE,tokenizer=tokenizer,strength=strength,do_cfg=do_cfg,cfg_scale=cfg_scale,sampler_name=sampler)

Image.fromarray(image)
```
